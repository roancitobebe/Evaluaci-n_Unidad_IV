# üî¢ Ejercicio 3: Verificaci√≥n de N√∫mero Primo (Valor - 5 pts)

### Descripci√≥n
Escribe un programa que determine si un n√∫mero ingresado por el usuario es primo. 
- Un n√∫mero primo es aquel que solo es divisible por 1 y por s√≠ mismo. 
- El programa debe utilizar un ciclo for para verificar la divisibilidad del n√∫mero por todos los enteros menores que √©l, optimizando el proceso hasta la ra√≠z cuadrada del n√∫mero.

Incluye estructuras condicionales para manejar casos especiales (n√∫meros menores o iguales a 1).

---

### ‚úÖ Salida esperada
El programa debe devolver un mensaje claro indicando si el n√∫mero es primo o no.

**Ejemplo:**
- Entrada: `numero = 7`  
- Salida: `Es primo`  

```python
# Aqu√≠ se resuelve el ejercic
import math
num = int(input("Ingrese su numero "))
sedivide = False

for i in range(2, int((math.sqrt(num))+1)):
    if num % i == 0:
        sedivide = True
        break
if sedivide:
    print(f"{num} no es primo")
else:
    print(f"{num} es primo") 
    
```
